```python
from etfpy import get_available_etfs_list, load_etf

def find_etfs_with_stock(stock_symbol):
    etfs_with_stock = []
    etfs = get_available_etfs_list()
    
    for etf_symbol in etfs:
        etf = load_etf(etf_symbol)
        holdings = etf.holdings
        
        if any(holding['Symbol'] == stock_symbol for holding in holdings):
            etfs_with_stock.append(etf_symbol)
    
    return etfs_with_stock

if __name__ == "__main__":
    stock_symbol = input("Enter the stock symbol: ").strip().upper()
    etfs = find_etfs_with_stock(stock_symbol)
    
    if etfs:
        print(f"ETFs containing {stock_symbol}:")
        for etf in etfs:
            print(etf)
    else:
        print(f"No ETFs found containing the stock symbol {stock_symbol}.")
```

```python
from etfpy import get_available_etfs_list, load_etf
from concurrent.futures import ThreadPoolExecutor, as_completed
import threading
import csv
import os

# Initialize a lock for thread-safe printing
print_lock = threading.Lock()

# Directory to store the CSV files
csv_directory = 'etf_holdings'

def ensure_directory_exists(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

def check_etf_for_stock(etf_symbol, stock_symbol, counter):
    etf = load_etf(etf_symbol)
    holdings = etf.holdings
    
    # Write holdings to CSV
    csv_filename = os.path.join(csv_directory, f"{etf_symbol}.csv")
    with open(csv_filename, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=holdings[0].keys())
        writer.writeheader()
        writer.writerows(holdings)
    
    with print_lock:
        print(f"Processed ETF {counter}: {etf_symbol} - Holdings written to {csv_filename}")
    
    return etf_symbol if any(holding['Symbol'] == stock_symbol for holding in holdings) else None

def find_etfs_with_stock(stock_symbol):
    etfs_with_stock = []
    etfs = get_available_etfs_list()
    total_etfs = len(etfs)
    counter = 0
    
    with ThreadPoolExecutor() as executor:
        futures = {executor.submit(check_etf_for_stock, etf, stock_symbol, i+1): etf for i, etf in enumerate(etfs)}
        
        for future in as_completed(futures):
            etf_symbol = future.result()
            if etf_symbol:
                etfs_with_stock.append(etf_symbol)
                print(etfs_with_stock)
    
    return etfs_with_stock

if __name__ == "__main__":
    stock_symbol = input("Enter the stock symbol: ").strip().upper()
    
    # Ensure the CSV directory exists
    ensure_directory_exists(csv_directory)
    
    etfs = find_etfs_with_stock(stock_symbol)
    
    if etfs:
        print(f"ETFs containing {stock_symbol}:")
        for etf in etfs:
            print(etf)
    else:
        print(f"No ETFs found containing the stock symbol {stock_symbol}.")
```
